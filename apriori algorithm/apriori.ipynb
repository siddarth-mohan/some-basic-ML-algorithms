```
#importing the necessary libraries
import numpy as np
import pandas as pd
from apyori import apriori
```

```
#reading the data from csv file
store_data = pd.read_csv('groceries.csv',error_bad_lines=False)
# head() function will show the first 5 rows of the table
store_data.head()
```

```
# to see the number of rows and columns
store_data.shape
```

```
#making a list of transactions where the transaction itself if a list of items purchased in that transaction
records=[]
for i in range (0,6105):
  records.append([str(store_data.values[i,j]) for j in range (0,4)])
```

```
records[0]
```

```
#removing nan values from the transactions
final_list = []
for i in range(len(records)):
  temp = []
  for x in records[i]:
    if str(x)!='nan':
      temp.append(x)
  final_list.append(temp)
```

```
final_list[0]
```

```
#applying apriori algorith to the final list
association_rules = apriori(final_list, min_support=0.002, min_confidence=0.3)
association_results= list(association_rules)
```

```
print(association_results[0])
```

```
#printing the association rules with support, confidence and lift
for item in association_results:

  pair=item[0]
  items = [x for x in pair]
  print("Rule: "+items[0] + " -> " + items[1])

  print("support: "+ str(item[1]))

  print("confidence: " + str((item[2][0][2])))
  print("Lift:  " + str(item[2][0][3]))
  print("================================")
```
